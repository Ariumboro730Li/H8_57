```python
import numpy as np
import pandas as pd

pd.set_option('display.max.columns',None)

url = "https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/Canada.xlsx?raw=true"
source = pd.read_excel(url, 
                       skiprows=range(20),
                       skipfooter=2,
                        sheet_name='Canada by Citizenship')

source.head()
```

```python
newIndex = source.set_index('OdName')
newIndex.head()
years = list(range(2000, 2014))
China = newIndex.loc["China",  years]
China.plot()
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
#print('mpl version:', mpl.__version__)

#plt.title("Albania")
#plt.ylabel("Total")
#plt.xlabel("Year")

Alban = newIndex.loc["India", years]
Alban.plot(kind="area", 
           figsize=(14,8),
           )
Alban.transpose().plot()
```

```python
source['Total'] = source.sum(axis=1)
newIndex = source.set_index('OdName')
top5     = newIndex.head(5)
top5     = top5[years].transpose()
top5
top5.index = top5.index.map(int)
top5.plot(kind='hist', figsize=(14,8))
plt.title('Top 5')
plt.ylabel('Imigrants')
plt.xlabel('Years')
plt.show()
```

```python
histo = newIndex.loc[['Denmark', 'Norway', 'Sweden'], years]
histo = histo.transpose()
histo = histo.head(10)
histo.plot(kind = 'hist', figsize=(10,6))

plt.title("Histogram of Immigration from Denmark, Norway, Sweden from 1980 - 2013")
plt.ylabel('Years')
plt.xlabel('Immigrants')
plt.show()
```
