```python
file_pol = 'Police_Department_Incidents_-_Previous_Year__2016_.csv';
df_incidents = pd.read_csv(file_pol)
print('Dataset downloaded and read into a pandas dataframe!')
```

```python
df_head = df_incidents.head(100)
df_head
```

```python
df_incidents.shape
```

```python
latitude = 37.77
longitude = -122.42

sanfran_map = folium.Map(location=[latitude, longitude], zoom_start=12)

# display the map of San Francisco
sanfran_map
```

```python
incidents = folium.map.FeatureGroup()

for lat, lng, in zip(df_head.Y, df_incidents.X):
    incidents.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=5, 
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )

sanfran_map.add_child(incidents)
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="location script")
location = geolocator.geocode("Bekasi")

latitude = location.latitude
longitude = location.longitude

semarang_map = folium.Map(location=[latitude, longitude], zoom_start=12)

# display the map of San Francisco
semarang_map.add_child(incidents)
```
